```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.io
import pywavefront as pwf
```

```python
f = scipy.io.loadmat("G:/Phd/Data/hand.mat")
X = np.array(f["X"])
y = np.array(f["Y"])
```

```python
print(X.shape)
print(y.shape)
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection="3d")
ax.scatter(X[:,0],X[:,1],X[:,2],marker="o",c=y,cmap='rainbow',s=5)
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')
plt.show()
```
